Correção exercicio Conversão de dolar para real

```python
##Recebe a cotação do dollar e o valor em reais
dolarcotacao = float(input("Coloque a cotação atual do dolar: "))
valoremreais = float(input("Coloque o valor em reais: "))
```

Correção exercicio comissão

```python
##Cálculo de comissão
precounitario = float(input("Coloque o preço unitário do produto: "))
quantidadedevendas = float(input("Coloque a quantidade vendida: "))
```

```python
##Calculo
valortotalvenda = precounitario * quantidadedevendas
comissao = valortotalvenda * 0.05
```

```python
##Resultado
print(f"Valor total da venda: R$`{valortotalvenda:.2f} ")
print(f"Comissão do vendedor: R$ {comissao:.2f}")
```

Exercicio Correção carro

```python
##Começo
custo_fabrica = float(input("Digite o custo de fábrica do carro (R$): "))
```

```python
##meio
impostos = custo_fabrica * 0.45
preco_com_impostos = custo_fabrica + impostos
distribuicao = preco_com_impostos * 0.28
custo_consumidor = preco_com_impostos + distribuicao
```

Exercicio correção hora extra

```python
##inicio
horas_trabalhadas = float(input("Quantas horas você trabalhou? "))
minutos_trabalhados = float(input("Quantos minutos você trabalhou? "))
valor_por_hora = float(input("Quanto você ganha por hora? R$ "))
```

```python
##meio
horas_decimais = horas_trabalhadas + (minutos_trabalhados / 60)
valor_hora_extra = valor_por_hora * 1.70
valor_total = horas_decimais * valor_hora_extra
```

```python
##fim
print(f"Você irá receber R$ {valor_total:.2f}")
```

Exercicio extra Simulação de Investimento em CDB com Juros Compostos

```python
##inicio
capital_investido = float(input("Qual o valor inicial do investimento? R$ "))
taxa_juros = float(input("Qual a taxa de juros anual (em %)? "))
anos = float(input("Qual o período de tempo em anos? "))
```

```python
##meio
taxa_juros_decimal = taxa_juros / 100
montante_final = capital_investido * (1 + taxa_juros_decimal) ** anos
```

```python
##fim
print(f"O montante final após {anos} anos é: R$ {montante_final:.2f}")
```

exercicio extra final restaurante

```python
##inicio
valor_conta = float(input("Qual o valor total da conta? R$ "))
gorjeta = float(input("Qual o valor da gorjeta? R$ "))
numero_pessoas = float(input("Quantas pessoas vão dividir a conta? "))
```

```python
##meio
taxa_servico = valor_conta * 0.10
valor_total = valor_conta + taxa_servico + gorjeta
valor_por_pessoa = valor_total / numero_pessoas
```

```python
print(f"\nTaxa de serviço (10%): R$ {taxa_servico:.2f}")
print(f"Valor total da conta (conta + taxa de serviço + gorjeta): R$ {valor_total:.2f}")
print(f"Valor por pessoa: R$ {valor_por_pessoa:.2f}")
```
